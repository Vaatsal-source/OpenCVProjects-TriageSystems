```python
##----------->>>>>>If,Else,Elif Statements 


#name = 'Vaatsal'
#if name =='Vaatsal':
#    print("Hello Vaatsal")
#else :
#    print("You are not Vaatsal!!")


##----------->>>>>Loops ( Done In The Basics)


#spam=0
#while spam<5:
#    print("The Value Is Still Less than 5")
#    spam=spam+1
#message='hello from here'
#for i in range(4):
#    print('the message for you is'+ message + str(i) +'times')
#for i in range (0,10,2):
#    print(i)
#for i in range(5,-1,-1):
#    print(i)


##----------->>>>>>>Import Statements


#import random 
#for i in range(5):
#    print(random.randint(1,10))
#import sys
#while True:
#    print('Type Exit To Exit')
#    response=input('>')
#    if(response=='Exit'):
#        sys.exit()
#    print('You Typed'+ response +'.' )


##----------->>>>>>>>> This is a guess the number game.


#import random
#secret_number = random.randint(1, 20)
#print('I am thinking of a number between 1 and 20.')
#
## Ask the player to guess 6 times.
#for guesses_taken in range(1, 7):
#    print('Take a guess.')
#    guess = int(input('>'))
#
#    if guess < secret_number:
#        print('Your guess is too low.')
#    elif guess > secret_number:
#        print('Your guess is too high.')
#    else:
#        break  # correct guess
#
#if guess == secret_number:
#    print('Good job! You got it in ' + str(guesses_taken) + ' guesses!')
#else:
#    print('Nope. The number was ' + str(secret_number))


##------------->>>>>>>> Rock Paper Scissor Game

#import random, sys
#
#print('ROCK, PAPER, SCISSORS')
#
#wins = 0
#losses = 0
#ties = 0
#
#while True:  # game loop
#    print('%s Wins, %s Losses, %s Ties' % (wins, losses, ties))
#    while True:  # input 
#        print('Enter your move: (r)ock (p)aper (s)cissors or (q)uit')
#        player_move = input('>')
#        if player_move == 'q':
#            sys.exit()  # Quit the game.
#        if player_move == 'r' or player_move == 'p' or player_move == 's':
#            break 
#        print('Type one of r, p, s, or q.')
#
#    # player choice:


#    if player_move == 'r':
#        print('ROCK versus...')
#    elif player_move == 'p':
#        print('PAPER versus...')
#    elif player_move == 's':
#        print('SCISSORS versus...')
#
#    # computer choice:
#    move_number = random.randint(1, 3)
#    if move_number == 1:
#        computer_move = 'r'
#        print('ROCK')
#    elif move_number == 2:
#        computer_move = 'p'
#        print('PAPER')
#    elif move_number == 3:
#        computer_move = 's'
#        print('SCISSORS')
#
#    # Display and record the win/loss/tie:


#    if player_move == computer_move:
#        print('It is a tie!')
#        ties = ties + 1
#    elif player_move == 'r' and computer_move == 's':
#        print('You win!')
#        wins = wins + 1
#    elif player_move == 'p' and computer_move == 'r':
#        print('You win!')
#        wins = wins + 1
#    elif player_move == 's' and computer_move == 'p':
#        print('You win!')
#        wins = wins + 1
#    elif player_move == 'r' and computer_move == 'p':
#        print('You lose!')
#        losses = losses + 1
#    elif player_move == 'p' and computer_move == 's':
#        print('You lose!')
#        losses = losses + 1
#    elif player_move == 's' and computer_move == 'r':
#        print('You lose!')
#        losses = losses + 1


##------------->>>>>>>>>>>Functions


#def hello():
#    print("Guten Morgen")
#    print("Guten Abend")
#    print("Guten Nacht")
#hello()

#   CALLBACK HELL 

#def a():
#    print('a() starts')
#    b()
#    d()
#    print('a() returns')
#
#def b():
#    print('b() starts')
#    c()
#    print('b() returns')
#
#def c():
#    print('c() starts')
#    print('c() returns')
#
#def d():
#    print('d() starts')
#    print('d() returns')
#
#a()


##--------->>>>>>>>>SCOPE DECLARATIONS


#def spam():
#   global eggs
#   eggs = 'spam'
#
#eggs = 'global'
#spam()
#print(eggs)  


##--------->>>>>>>ERROR HANDLING 


#def spam(divide_by):
#    try:
#        return 42 / divide_by
#    except ZeroDivisionError:
#        print('Error: Invalid argument.')
#
#print(spam(2))
#print(spam(12))
#print(spam(0))
#print(spam(1))


##--------->>>>>>>>>ZIGZAG PATTERN

#import time, sys
#indent = 0  
#indent_increasing = True  
#
#try:
#    while True:  
#        print(' ' * indent, end='')
#        print('********')
#        time.sleep(0.1) 
#
#        if indent_increasing:
#            # Increasing the number of spaces:
#            indent = indent + 1
#            if indent == 20:
#                # Changing direction:
#                indent_increasing = False
#        else:
#            # Decrease the number of spaces:
#            indent = indent - 1
#            if indent == 0:
#                # Again Changing direction:
#                indent_increasing = True
#except KeyboardInterrupt:
#    sys.exit()


#----------->>>>>>>>>>>>>>Exception Handling 


#def box_print(symbol, width, height):
#    if len(symbol) != 1:
#       raise Exception('Symbol must be a single character string.')
#    if width <= 2:
#       raise Exception('Width must be greater than 2.')
#    if height <= 2:
#       raise Exception('Height must be greater than 2.')
#
#    print(symbol * width)
#    for i in range(height - 2):
#        print(symbol + (' ' * (width - 2)) + symbol)
#    print(symbol * width)
#
#try:
#    box_print('*', 4, 4)
#    box_print('O', 20, 5)
#    box_print('x', 1, 3)
#    box_print('ZZ', 3, 3)
#except Exception as err:
#    print('An exception happened: ' + str(err))
#try:
#    box_print('ZZ', 3, 3)
#except Exception as err:
#    print('An exception happened: ' + str(err))


##---------------->>>>>>>>>>>>>>>>Logging 


import logging
logging.basicConfig(level=logging.DEBUG, format='%(asctime)s -  %(levelname)s -  %(message)s')
logging.debug('Start of program')

def factorial(n):
    logging.debug('Start of factorial(' + str(n) + ')')
    total = 1
    for i in range(1,n + 1):
        total *= i
        logging.debug('i is ' + str(i) + ', total is ' + str(total))
    logging.debug('End of factorial(' + str(n) + ')')
    return total

print(factorial(5))
logging.debug('End of program')



```
